```python
saldo = 1000
historico_depositos = []
historico_saques = []

def menu(saldo):
    print('\n===Menu Caixa Eletrônico===\n1-Depositar\n2-Sacar\n3-Ver saldo\n4-Ver histórico de transações\n5-Sair')
    opcao = int(input('Escolha uma opção: '))
    if opcao == 1:
        saldo = depositar(saldo)
    elif opcao == 2:
        saldo = sacar(saldo)
    elif opcao == 3:
        ver_saldo(saldo)
    elif opcao == 4:
        historico()
    elif opcao == 5:
        print('Obrigado por usar o Caixa Eletrônico!')
        return exit()
    else:
        print('Opção inválida!')
    menu(saldo)

def depositar(saldo):
    valor = float(input('Digite o valor a ser depositado: '))
    historico_depositos.append(valor)
    print(f'\nDepósito de R${valor} realizado com sucesso!')
    saldo += valor
    return saldo

def historico():
    print('===Histórico de transações===\n')
    print('Depósitos:')
    if historico_depositos == [] and historico_saques:
        print('Nenhuma transação realizada')
    else:
      for valor in historico_depositos:
          print(f'R${valor}')
      print('Saques:')
      for valor in historico_saques:
          print(f'R${valor}')
    return

def sacar(saldo):
    valor = float(input('Digite o valor a ser sacado: '))
    if valor > saldo:
        print('Saldo insuficiente!')
        return saldo
    else:
        saldo -= valor
        historico_saques.append(valor)
        print(f'Saque de R${valor} realizado com sucesso!')
        return saldo

def ver_saldo(saldo):
    print(f'\nSeu saldo é de R${saldo}')
    return saldo

menu(saldo)
```
